# Waze app: user churn analysis and prediction model
Part 1: exploratory data analysis (EDA)

```python
# import packages and libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset into a dataframe
df = pd.read_csv("waze_dataset.csv")
```

```python
df.head(10)
```

The data is already in a structured format. Each row represents a user

```python
df.info()
```

There are 14,999 rows and 13 columns. 700 rows have label missing. Other variables have no missing values.

Is there a difference between the summary statistics of the 700 rows with missing labels and the rows that are not missing any values?


```python
null_df = df[df["label"].isnull()]

null_df.describe()
```

```python
not_null_df = df[-df["label"].isnull()]

not_null_df.describe()
```

No  discernible differences are found: the means and standard deviations are fairly consistent between the two groups.

Next, we may want to check how many iPhone users vs Android users have null values.

```python
null_df["device"].value_counts()
```

```python
null_df["device"].value_counts(normalize=True)
```

On a total of 700 rows with null values, 447 are iPhone users (64%) and 253 are Android users (36%).

```python
# calculate the percentage of iphone vs android users in the whole dataset
df["device"].value_counts(normalize=True)
```

By looking at the whole dataset, we find that the percentage of iPhone users vs Android users is consistent with the percentage found when analyzing only the missing values for each device. Therefore, we do not see any obvious pattern to the missing values.

Next, we look at the counts and percentages of users who churned vs those who were retained.

```python
df["label"].value_counts()
```

```python
df["label"].value_counts(normalize=True)
```

The dataset contains 11763 (82%) retained users and 2536 (18%) churned users.

Next, we calculate the median values by label (calculating the mean would be subject to the influence of outliers, while the median represents the middle value of the distribution regardless of any outlying values).

```python
df.groupby("label").median(numeric_only=True)
```

By comparing the medians of each variable for churned and retained users, we find that:

*   churned users performed an average of 3 more drives in the last month compared to the number of drives of retained users.
*   retained users used the app more than twice as many days as churned users.
*   the median churned user drove ~200 more km and for ~2.5 more hours than the median retained user.
*   churned users had more drives in fewer days (an average of 50 drives in 6 days) compared to retained users (with an average of 47 drives in 16 days).

This suggests that churned users had more and longer drives compared than the average retained user.


Based on the above considerations, we want to find for both groups:


*   the median kilometers per drive
*   the median kilometers per driving day
*   the median drives per driving day




```python
# calculate the median km per drive in the last month for both retained and churned users
median_by_label = df.groupby("label").median(numeric_only=True)

print('Median kilometers per drive:')
median_by_label["driven_km_drives"] / median_by_label["drives"]
```

We can see that the median user, whether churned or retained, drove an average of 73 km/drive.

```python
print('Median kilometers per driving day:')
median_by_label["driven_km_drives"] / median_by_label["driving_days"]
```

```python
print('Median drives per driving day:')
median_by_label["drives"] / median_by_label["driving_days"]
```

We see a significant difference in the kilometers per driving day of the median churned user (608km), compared to the kilometers per day of the median retained user (247km). Similarly, the median churned user had a significantly higher number of drives per day compared to the median retained user.

```python
# calculate how many users churned by device type
df.groupby(["label", "device"]).size()
```

```python
# calculate the percentage of how many users churned by device type
df.groupby("label")["device"].value_counts(normalize=True)
```

As we can see, the ratio of iPhone users and Android users is consistent between the churned group and the retained group. Therefore, nothing suggests that churn is correlated to the device used.
